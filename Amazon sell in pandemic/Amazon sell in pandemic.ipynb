```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("Amazon_sell_in_pandemic.csv")
```

```python
data['year']=pd.DatetimeIndex(data['date first available']).year
data
```

```python
pandemic_data=data[data['year']==2020]
pandemic_data
```

# Histogram

```python
sns.distplot(a=data.year,kde=False,color="r")
```

```python
data.describe()
```

# CountPlot

```python
plt.figure(figsize=(5,40))
sns.countplot(y=pandemic_data['brand name'],palette="Set2")
```

```python
df=pd.value_counts(pandemic_data['brand name'])
df=df.to_frame()
raw_data={
    "Brand Name":df.index,
    "Count":df['brand name']
}

df=pd.DataFrame(raw_data,columns=['Count'])
df
```

## From the above countplot we can conclude that NV Mart and TrendzOn products are the most popular product brands in pandemic

# Pie Plot

```python
pie=df.plot.pie(y='Count', figsize=(30, 10));
```
