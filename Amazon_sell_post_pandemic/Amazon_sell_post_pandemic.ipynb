```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("amazon_sell_post_pandemic.csv")
```

```python
data['month']=pd.DatetimeIndex(data=data['date']).month
```

```python
data.head()
```

## Checking for Missing values and outliers

```python
data.isnull().sum()
```

#### There are no missing values in the dataset

# Independent and dependent variables

#### Independent variables:- asin,date 

#### Dependent variable:- Product Name

# CountPlot

```python
plt.figure(figsize=(5,40))
sns.countplot(y=data['product name'],palette="Set3")
```

```python
df=pd.value_counts(data['product name'])
df=df.to_frame()
raw_data={
    "product Name":df.index,
    "Count":df['product name']
}

df=pd.DataFrame(raw_data,columns=['Count'])
df
```

### From above plot and dataset we can conclude that Dettol-Disinfectant-Cleaner-Home-Fresh is the most popular product in post pandemic period

# Pie Plot

```python
pie=df.plot.pie(y='Count', figsize=(30, 10));
```

## Histogram

```python
sns.distplot(a=data['month'],color="g")
```

### We can see that most products are sold in 8 month i.e (August Month)
