```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
data=pd.read_excel("AQI2020HYD.xlsx")
```

```python
data.columns
data.index=data['Location']
data=data.iloc[:,1:]
```

```python
data.head()
```

```python
data.describe()
```

```python
avg_aqi=data['Avg'];
avg_aqi.describe()
```

# Dependent and independent variables

#### Dependent variable:       Avg AQI(Air Quality Index) in the year 2020

#### Independent Variable :    Location of stations in Hyderabad

# Univariate analysis

# Histogram

```python
sns.distplot(a=avg_aqi,bins=6,color="tab:cyan")
```

# Bar Plot

```python
plt.figure(figsize=(10,5))
sns.barplot(y=data.index,x=avg_aqi)
```

# Pie Plot

```python
plot = data.plot.pie(y='Avg', figsize=(30, 10))
```

# Box Plot

```python
sns.boxplot(x=avg_aqi)
```

### box plot confirms that there are not outliers in the data

## Meteorological factors that correlate with the air pollution 

GOOD  (0-50)	Minimal impact	
SATISFACTORY(51-100)	Minor breathing discomfort to sensitive people	
MODERATE (101-200)	Breathing discomfort to the people with lungs, asthma and heart disease	
POOR (201-300)	Breathing discomfort to most people on prolonged exposure	
VERY POOR (301-400)	Respiratory illness to people on prolonged exposure	
SEVERE  (>400)	Affects healthy people and seriously impacts those with existing diseases	


```python
max_aqi_area=data[data['Avg']==data['Avg'].max()]
max_aqi_area
```

### From above we can conclude that Jeedimetla, Industrial Association building area has the highest AQI (but it has satisfactory AQI).

```python
min_aqi_area=data[data['Avg']==data['Avg'].min()]
min_aqi_area
```

### From above we can conclude that Jeedimetla, Industrial Association building area has the lowest AQI
